## 1.1 Processing

第一次使用Pytorch，只能摸索着来。预处理的步骤如下：

- 读取data，resize
- 定义好transformer，用于数据增广，标准化，注意test中并不需要做数据增广，因而需要定义两个不同的transformer

`````python
# read data and resize
# data_path = r'./food-11'
data_path = r'/content/food-11 - 副本'
# data_path = r'E:\p\HungyiLee\ML Lecture\week4\food-11'
`````

`````python
def read_data(path, with_label = True):
    width = 128
    channel = 3
    image_dirs = sorted(os.listdir(path))
    x = np.zeros((len(image_dirs), width, width, channel),  dtype=np.uint8)
    y = np.zeros((len(image_dirs),),  dtype=np.uint8)
    for i,img_dir in enumerate(image_dirs):
        img = cv2.imread(os.path.join(path,img_dir))
        x[i] = cv2.resize(img, (width, width))
        if with_label:
            y[i] = int(img_dir.split('_')[0])
    if with_label:
        return x,y
    return x
`````

`````python
train_x, train_y = read_data(os.path.join(data_path, 'training'))
val_x, val_y = read_data(os.path.join(data_path, 'validation'))
test_x= read_data(os.path.join(data_path, 'testing'), False)
print('train shape {}'.format(train_x.shape))
print(train_y[-1], train_x[-1, 1, 1])
print('val_shape {}'.format(val_x.shape))
print(val_y[-1], val_x[-1, 1, 1])
print('test_shape{}'.format(test_x.shape))
print(test_x[-1,1,1])
`````

**Dataset Operation of Torch**

Torch中数据集相关API主要是三个部分，Dataset，transform和DataLoader，其中transform部分其实可以是任何需要对数据集进行处理的对象。

这边的设计明显是用了代理，DataLoader是对数据集进行迭代的的类，它操作Dataset，负责批量迭代生成数据等，所以我们必须继承Dataset，为数据集实现当前Dataset子类，主要是覆盖长度获取函数，访问函数。

Dataset另一个作用是在操作的时候对数据进行预处理，可以实现数据或io的懒加载。


`````python
# # define transformer
# train_transform = transforms.Compose([
#     transforms.ToPILImage(), 
#     transforms.ToTensor(), # normalization
# ])
# test_transform = transforms.Compose([
#     transforms.ToPILImage(), 
#     transforms.ToTensor(), # normalization
# ])

#training 時做 data augmentation
train_transform = transforms.Compose([
    transforms.ToPILImage(),
    transforms.RandomHorizontalFlip(), #隨機將圖片水平翻轉
    transforms.RandomRotation(15), #隨機旋轉圖片
    transforms.ToTensor(), #將圖片轉成 Tensor，並把數值normalize到[0,1](data normalization)
])
#testing 時不需做 data augmentation
test_transform = transforms.Compose([
    transforms.ToPILImage(),                                    
    transforms.ToTensor(),
])
`````

`````python
# extend Dataset
class ImgDataset(Dataset):
    def __init__(self, x, y = None, transform = None):
        self.x = x
        self.y = y
        if y is not None:
            self.y = torch.LongTensor(y) # label required to be LongTensor
        self.transform = transform
    def __len__(self):
        return len(self.x)
    def __getitem__(self, index):
        X = self.x[index]
        if self.transform is not None:
            X = self.transform(X)
        if self.y is not None:
            return X, self.y[index]
        else:
            return X
`````

`````python
train_set = ImgDataset(train_x, train_y, train_transform)
val_set = ImgDataset(val_x, val_y, test_transform)
# test_set = ImgDataset(test_x)
print("train size {}".format(len(train_set)))
xx,yy = train_set[1]
print("with normalization, train_x shape is : {}      value is : {}      train_y is {}".format(xx.shape,xx[:, 1, 1],yy))
print("val size {}".format(len(val_set)))
xx,yy = val_set[1]
print("with normalization, val_x shape is : {}      value is : {}      val_y is {}".format(xx.shape,xx[:, 1, 1],yy))
# print("test size {}".format(len(test_set)))
# print("without normalization {}".format(test_set[-1,1,1]))
`````

## 1.2 Building Model



`````python
class Classifier(nn.Module):
    def __init__(self):
        super(Classifier, self).__init__()
        #torch.nn.Conv2d(in_channels, out_channels, kernel_size, stride, padding)
        #torch.nn.MaxPool2d(kernel_size, stride, padding)
        #input 維度 [3, 128, 128]
        self.cnn = nn.Sequential(
            nn.Conv2d(3, 64, 3, 1, 1),  # [64, 128, 128]
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(2, 2, 0),      # [64, 64, 64]

            nn.Conv2d(64, 128, 3, 1, 1), # [128, 64, 64]
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.MaxPool2d(2, 2, 0),      # [128, 32, 32]

            nn.Conv2d(128, 256, 3, 1, 1), # [256, 32, 32]
            nn.BatchNorm2d(256),
            nn.ReLU(),
            nn.MaxPool2d(2, 2, 0),      # [256, 16, 16]

            nn.Conv2d(256, 512, 3, 1, 1), # [512, 16, 16]
            nn.BatchNorm2d(512),
            nn.ReLU(),
            nn.MaxPool2d(2, 2, 0),       # [512, 8, 8]
            
            nn.Conv2d(512, 512, 3, 1, 1), # [512, 8, 8]
            nn.BatchNorm2d(512),
            nn.ReLU(),
            nn.MaxPool2d(2, 2, 0),       # [512, 4, 4]
        )
        self.fc = nn.Sequential(
            nn.Linear(512*4*4, 1024),
            nn.ReLU(),
            nn.Linear(1024, 512),
            nn.ReLU(),
            nn.Linear(512, 11)
        )

    def forward(self, x):
        out = self.cnn(x)
        out = out.view(out.size()[0], -1)
        return self.fc(out)
`````

## 1.3 Training

Pytorch的训练需要手动设置的地方较多。Batch迭代算法大致如下：

````python
for epoch
    for x_batch in train_data
        # forward
        pred_batch <- model.predict(x_batch)
        loss_batch <- loss_function(pred_batch, y_batch)

        # backward
        optim: gradient descent

        # record log
        acc += calculate batch acc (pred_batch, y_batch)
        loss += loss_batch

    calculate val loss & acc
````

`````python
model = Classifier().cuda()
loss_func = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr = 0.001) # optim take charge of gradient descent
num_epoch = 30
# log
logs ={'time':[], 'train_loss': [], 'train_acc':[], 'val_loss':[], 'val_acc': []}
for epoch in range(num_epoch):
    epoch_start_time = time.time() # 记录时间
    train_acc = 0.0 # 注意，这里的acc和loss相关的值，其实都是计算累积loss和累计正确预测个数，没有做均值处理
    train_loss = 0.0
    
    model.train() # 设置训练状态，由于dropout，batch_normalization策略在train和test时不同
    for i, (x_batch,y_batch) in enumerate(train_loader): # train loder返回batch size的x，y
        # forward
        optimizer.zero_grad() # 由于每次只计算一个batch，所以optim的默认策略是不断累加grad
        pred_batch = model(x_batch.cuda()) # torch的特点，需要将数据设置到gpu上
        loss_batch = loss_func(pred_batch, y_batch.cuda())
        # backward
        loss_batch.backward()
        optimizer.step()
        # log
        train_acc += np.sum(np.argmax(pred_batch.cpu().data.numpy(), axis=1) == y_batch.numpy()) # 为什么要搬回cpu
        train_loss += loss_batch
    
    # calculate val acc and loss
    model.eval()
    val_acc = 0.0
    val_loss = 0.0
    with torch.no_grad():
        for i,(x_batch, y_batch) in enumerate(val_loader):
            pred_batch = model(x_batch.cuda())
            loss_batch = loss_func(pred_batch, y_batch.cuda())
            val_acc +=  np.sum(np.argmax(pred_batch.cpu().data.numpy(), axis=1) == y_batch.numpy()) 
            val_loss += loss_batch
    # print log
    logs['time'].append(time.time()-epoch_start_time)
    logs['train_acc'].append(train_acc/train_set.__len__())
    logs['train_loss'].append(train_loss/train_set.__len__())
    logs['val_acc'].append(val_acc/val_set.__len__())
    logs['val_loss'].append(val_loss/val_set.__len__()) # CrossEntropy包含softmax，但并没有求均值
    
    print('[%03d/%03d] %2.2f sec(s) Train Acc: %3.6f Loss: %3.6f | Val Acc: %3.6f loss: %3.6f' % \
            (epoch + 1, num_epoch,  logs['time'][-1], \
            logs['train_acc'][-1] , logs['train_loss'][-1], logs['val_acc'][-1], logs['val_loss'][-1]))
    
`````

`````python
# 平滑函数
def smooth_curve(points, factor = 0.9):
    smoothed_points = []
    for point in points:
        if smoothed_points:
            prev = smoothed_points[-1]
            smoothed_points.append(prev * factor + point * (1 - factor))
        else:
            smoothed_points.append(point)
    return smoothed_points

from matplotlib import pyplot as plt
plt.plot(logs['train_loss'], label = 'train loss')
plt.plot(smooth_curve(logs['val_loss'], 0), label = 'val loss')
plt.legend()
plt.show()
plt.plot(logs['train_acc'], label = 'train acc')
plt.plot(logs['val_acc'], label = 'val acc')
plt.legend()
plt.show()
`````

`````python
class Classifier(nn.Module):
    def __init__(self):
        super(Classifier, self).__init__()
        #torch.nn.Conv2d(in_channels, out_channels, kernel_size, stride, padding)
        #torch.nn.MaxPool2d(kernel_size, stride, padding)
        #input 維度 [3, 128, 128]
        self.fc = nn.Sequential(
            nn.Linear(3*128*128, 224),
            nn.ReLU(),
            nn.Linear(224, 128),
            nn.ReLU(),
            nn.Linear(128, 256),
            nn.ReLU(),
            nn.Linear(256, 512),
            nn.ReLU(),
            nn.Linear(512, 512),
            nn.ReLU(),
            nn.Linear(512, 1024),
            nn.ReLU(),  
            nn.Linear(1024, 512),
            nn.ReLU(),                      
            nn.Linear(512, 11)
        )

    def forward(self, x):
        out = x.view(x.size()[0], -1)# 直接flatten
        return self.fc(out)
`````

### 1.5.5  Data Augmentation

图像的数据增广很有代表性，核心是：不让模型看到相同的图片。

motivation是，比如说我们看一只猫，不管近看远看，左看右看它都应该是一只猫，即结果不应该受到图片大小，方向，角度的影响。

图像数据增广的基本思想是，在训练模型输入图片时对角度、大小、方向等进行随机变化，这种做法还可以让模型每次都看见不同的图片，不容易过拟合。

notes: 本来这里应该研究normalization，但是我前面实验中默认做了normalization，所以这里就不重做了，图像的normalization比较简单，结果可以预期——最大的作用是数据归一化，使得收敛变得容易，数据绝对值变小，不容易溢出。



`````python
#training 時做 data augmentation
train_transform = transforms.Compose([
    transforms.ToPILImage(),
    transforms.RandomHorizontalFlip(), #隨機將圖片水平翻轉
    transforms.RandomRotation(15), #隨機旋轉圖片
    transforms.ToTensor(), #將圖片轉成 Tensor，並把數值normalize到[0,1](data normalization)
])
#testing 時不需做 data augmentation
test_transform = transforms.Compose([
    transforms.ToPILImage(),                                    
    transforms.ToTensor(),
])
class ImgDataset(Dataset):
    def __init__(self, x, y=None, transform=None):
        self.x = x
        # label is required to be a LongTensor
        self.y = y
        if y is not None:
            self.y = torch.LongTensor(y)
        self.transform = transform
    def __len__(self):
        return len(self.x)
    def __getitem__(self, index):
        X = self.x[index]
        if self.transform is not None:
            X = self.transform(X)
        if self.y is not None:
            Y = self.y[index]
            return X, Y
        else:
            return X
`````

`````python
batch_size = 128
train_set = ImgDataset(train_x, train_y, train_transform)
val_set = ImgDataset(val_x, val_y, test_transform)
train_loader = DataLoader(train_set, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_set, batch_size=batch_size, shuffle=False)
`````

`````python
class Classifier(nn.Module):
    def __init__(self):
        super(Classifier, self).__init__()
        #torch.nn.Conv2d(in_channels, out_channels, kernel_size, stride, padding)
        #torch.nn.MaxPool2d(kernel_size, stride, padding)
        #input 維度 [3, 128, 128]
        self.cnn = nn.Sequential(
            nn.Conv2d(3, 64, 3, 1, 1),  # [64, 128, 128]
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(2, 2, 0),      # [64, 64, 64]

            nn.Conv2d(64, 128, 3, 1, 1), # [128, 64, 64]
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.MaxPool2d(2, 2, 0),      # [128, 32, 32]

            nn.Conv2d(128, 256, 3, 1, 1), # [256, 32, 32]
            nn.BatchNorm2d(256),
            nn.ReLU(),
            nn.MaxPool2d(2, 2, 0),      # [256, 16, 16]

            nn.Conv2d(256, 512, 3, 1, 1), # [512, 16, 16]
            nn.BatchNorm2d(512),
            nn.ReLU(),
            nn.MaxPool2d(2, 2, 0),       # [512, 8, 8]
            
            nn.Conv2d(512, 512, 3, 1, 1), # [512, 8, 8]
            nn.BatchNorm2d(512),
            nn.ReLU(),
            nn.MaxPool2d(2, 2, 0),       # [512, 4, 4]
        )
        self.fc = nn.Sequential(
            nn.Linear(512*4*4, 1024),
            nn.ReLU(),
            nn.Linear(1024, 512),
            nn.ReLU(),
            nn.Linear(512, 11)
        )

    def forward(self, x):
        out = self.cnn(x)
        out = out.view(out.size()[0], -1)
        return self.fc(out)
`````

`````python
model = Classifier().cuda()
loss_func = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr = 0.001) # optim take charge of gradient descent
num_epoch = 30
# log
logs ={'time':[], 'train_loss': [], 'train_acc':[], 'val_loss':[], 'val_acc': []}
for epoch in range(num_epoch):
    epoch_start_time = time.time() # 记录时间
    train_acc = 0.0 # 注意，这里的acc和loss相关的值，其实都是计算累积loss和累计正确预测个数，没有做均值处理
    train_loss = 0.0
    
    model.train() # 设置训练状态，由于dropout，batch_normalization策略在train和test时不同
    for i, (x_batch,y_batch) in enumerate(train_loader): # train loder返回batch size的x，y
        # forward
        optimizer.zero_grad() # 由于每次只计算一个batch，所以optim的默认策略是不断累加grad
        pred_batch = model(x_batch.cuda()) # torch的特点，需要将数据设置到gpu上
        loss_batch = loss_func(pred_batch, y_batch.cuda())
        # backward
        loss_batch.backward()
        optimizer.step()
        # log
        train_acc += np.sum(np.argmax(pred_batch.cpu().data.numpy(), axis=1) == y_batch.numpy()) # 为什么要搬回cpu
        train_loss += loss_batch
    
    # calculate val acc and loss
    model.eval()
    val_acc = 0.0
    val_loss = 0.0
    with torch.no_grad():
        for i,(x_batch, y_batch) in enumerate(val_loader):
            pred_batch = model(x_batch.cuda())
            loss_batch = loss_func(pred_batch, y_batch.cuda())
            val_acc +=  np.sum(np.argmax(pred_batch.cpu().data.numpy(), axis=1) == y_batch.numpy()) 
            val_loss += loss_batch
    # print log
    logs['time'].append(time.time()-epoch_start_time)
    logs['train_acc'].append(train_acc/train_set.__len__())
    logs['train_loss'].append(train_loss/train_set.__len__())
    logs['val_acc'].append(val_acc/val_set.__len__())
    logs['val_loss'].append(val_loss/val_set.__len__()) # CrossEntropy包含softmax，但并没有求均值
    
    print('[%03d/%03d] %2.2f sec(s) Train Acc: %3.6f Loss: %3.6f | Val Acc: %3.6f loss: %3.6f' % \
            (epoch + 1, num_epoch,  logs['time'][-1], \
            logs['train_acc'][-1] , logs['train_loss'][-1], logs['val_acc'][-1], logs['val_loss'][-1]))
    
`````

`````python
class Classifier(nn.Module):
    def __init__(self):
        super(Classifier, self).__init__()
        #torch.nn.Conv2d(in_channels, out_channels, kernel_size, stride, padding)
        #torch.nn.MaxPool2d(kernel_size, stride, padding)
        #input 維度 [3, 128, 128]
        self.cnn = nn.Sequential(
            nn.Conv2d(3, 64, 3, 1, 1),  # [64, 128, 128]
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(2, 2, 0),      # [64, 64, 64]

            nn.Conv2d(64, 128, 3, 1, 1), # [128, 64, 64]
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.MaxPool2d(2, 2, 0),      # [128, 32, 32]

            nn.Conv2d(128, 256, 3, 1, 1), # [256, 32, 32]
            nn.BatchNorm2d(256),
            nn.ReLU(),
            nn.MaxPool2d(2, 2, 0),      # [256, 16, 16]

            nn.Conv2d(256, 512, 3, 1, 1), # [512, 16, 16]
            nn.BatchNorm2d(512),
            nn.ReLU(),
            nn.MaxPool2d(2, 2, 0),       # [512, 8, 8]
            
            nn.Conv2d(512, 512, 3, 1, 1), # [512, 8, 8]
            nn.BatchNorm2d(512),
            nn.ReLU(),
            nn.MaxPool2d(2, 2, 0),       # [512, 4, 4]
        )
        self.fc = nn.Sequential(
            nn.Dropout(0.5),
            nn.Linear(512*4*4, 1024),
            nn.ReLU(),
            nn.Linear(1024, 512),
            nn.ReLU(),
            nn.Linear(512, 11)
        )

    def forward(self, x):
        out = self.cnn(x)
        out = out.view(out.size()[0], -1)
        return self.fc(out)
`````

`````python
model = Classifier().cuda()
loss_func = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr = 0.001) # optim take charge of gradient descent
num_epoch = 50
# log
logs ={'time':[], 'train_loss': [], 'train_acc':[], 'val_loss':[], 'val_acc': []}
for epoch in range(num_epoch):
    epoch_start_time = time.time() # 记录时间
    train_acc = 0.0 # 注意，这里的acc和loss相关的值，其实都是计算累积loss和累计正确预测个数，没有做均值处理
    train_loss = 0.0
    
    model.train() # 设置训练状态，由于dropout，batch_normalization策略在train和test时不同
    for i, (x_batch,y_batch) in enumerate(train_loader): # train loder返回batch size的x，y
        # forward
        optimizer.zero_grad() # 由于每次只计算一个batch，所以optim的默认策略是不断累加grad
        pred_batch = model(x_batch.cuda()) # torch的特点，需要将数据设置到gpu上
        loss_batch = loss_func(pred_batch, y_batch.cuda())
        # backward
        loss_batch.backward()
        optimizer.step()
        # log
        train_acc += np.sum(np.argmax(pred_batch.cpu().data.numpy(), axis=1) == y_batch.numpy()) # 为什么要搬回cpu
        train_loss += loss_batch
    
    # calculate val acc and loss
    model.eval()
    val_acc = 0.0
    val_loss = 0.0
    with torch.no_grad():
        for i,(x_batch, y_batch) in enumerate(val_loader):
            pred_batch = model(x_batch.cuda())
            loss_batch = loss_func(pred_batch, y_batch.cuda())
            val_acc +=  np.sum(np.argmax(pred_batch.cpu().data.numpy(), axis=1) == y_batch.numpy()) 
            val_loss += loss_batch
    # print log
    logs['time'].append(time.time()-epoch_start_time)
    logs['train_acc'].append(train_acc/train_set.__len__())
    logs['train_loss'].append(train_loss/train_set.__len__())
    logs['val_acc'].append(val_acc/val_set.__len__())
    logs['val_loss'].append(val_loss/val_set.__len__()) # CrossEntropy包含softmax，但并没有求均值
    
    print('[%03d/%03d] %2.2f sec(s) Train Acc: %3.6f Loss: %3.6f | Val Acc: %3.6f loss: %3.6f' % \
            (epoch + 1, num_epoch,  logs['time'][-1], \
            logs['train_acc'][-1] , logs['train_loss'][-1], logs['val_acc'][-1], logs['val_loss'][-1]))
    
`````

### 1.5.9 Resnet

使用残差网络。

残差网络Resnet-18结构并不复杂，其中连接弧线捷径的两个conv是一个ResidualBlock，其实仔细看左侧图片结构就会发现，所有block都是相同的配置：

- 进入block时，进行downsample，类似maxpooling的作用，
- 在block内部stride，padding，filter size都是一样的。

所以可以利用循环来完成网络的构建。

resnet如果说，还有什么一开始看着很晕的，大概就是stride的计算。公式如下，但其实，主要的注意力放在stride上就好了，但它为2时，那通道就会/2，这也是resnet的基本操作。
 
$$n_{\left(L\right)}=\frac{n_{\left(L-1\right)} + 2p-f}{s}-1$$

![替代文字](https://upload-images.jianshu.io/upload_images/15074510-ef99906b7dbbd4fd.png)

关于Resnet的有一个很有意思的残差学习理论。

> 如果一个浅层模型学得到，那么深层模型也应该没有问题，因为，后面的模型就算什么都不做，它也就等于浅层模型。或者，后面的模型只需要复制前面模型的东西，那么中间的，什么都不做，至少也是个浅层模型

**结论**

因为模型太大了，只能使用64x64x3的size，最终kaggle score是77。模型足够强了，看来还是得从数据预处理下手。

`````python
import torch
import torch.nn as nn
import torch.nn.functional as F

class ResidualBlock(nn.Module):
    def __init__(self, inchannel, outchannel, stride=1):
        super(ResidualBlock, self).__init__()
        self.left = nn.Sequential( # baise 固定设置为 False
            nn.Conv2d(inchannel, outchannel, kernel_size=3, stride=stride, padding=1, bias=False),
            nn.BatchNorm2d(outchannel), # 进入block配置都相同，除了stride需要注意以外
            nn.ReLU(inplace=True), # 节省内存用的？
            nn.Conv2d(outchannel, outchannel, kernel_size=3, stride=1, padding=1, bias=False),
            nn.BatchNorm2d(outchannel)
        )
        self.shortcut = nn.Sequential()
        # 决定是否downsample 原因是 ks = 1, stride = 2 
        # stride != 1 和 inchannel != outchannel是同时成立的
        if stride != 1 or inchannel != outchannel:
            self.shortcut = nn.Sequential( # 残差网络中没有使用maxpooling而是使用downsample
                nn.Conv2d(inchannel, outchannel, kernel_size=1, stride=stride, bias=False),
                nn.BatchNorm2d(outchannel)
            )

    def forward(self, x):
        out = self.left(x)
        out += self.shortcut(x)
        out = F.relu(out)
        return out

class ResNet(nn.Module):
    def __init__(self, ResidualBlock, num_classes=11):
        super(ResNet, self).__init__()
        self.inchannel = 64
        self.conv1 = nn.Sequential(
            nn.Conv2d(3, 64, kernel_size=3, stride=1, padding=1, bias=False),
            nn.BatchNorm2d(64),
            nn.ReLU(),
        )
        self.layer1 = self.make_layer(ResidualBlock, 64,  2, stride=1)
        # stride控制/2，其实只有第一次进入block时/2
        self.layer2 = self.make_layer(ResidualBlock, 128, 2, stride=2)
        self.layer3 = self.make_layer(ResidualBlock, 256, 2, stride=2)
        self.layer4 = self.make_layer(ResidualBlock, 512, 2, stride=2)
        self.fc = nn.Linear(2048, num_classes)

    def make_layer(self, block, channels, num_blocks, stride):
        strides = [stride] + [1] * (num_blocks - 1)   #strides=[2,1,1,1……]
        layers = []
        for stride in strides: # inchannel 和 outchannel交替
            layers.append(block(self.inchannel, channels, stride))
            self.inchannel = channels
        return nn.Sequential(*layers)

    def forward(self, x):
        # print(x.shape)
        out = self.conv1(x)
        # print(out.shape)
        out = self.layer1(out)
        # print(out.shape)
        out = self.layer2(out)
        # print(out.shape)
        out = self.layer3(out)
        # print(out.shape)
        out = self.layer4(out)
        # print(out.shape)
        out = F.avg_pool2d(out, 4)
        # print(out.shape)
        out = out.view(out.size(0), -1)
        # print(out.shape)
        out = self.fc(out)
        return out
`````
