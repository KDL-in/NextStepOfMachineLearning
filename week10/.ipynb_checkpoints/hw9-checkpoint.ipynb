## Preprocess Data

```python
'''
计算mean std
'''
def cal_mean_std(data):
  data = data / 255.0
  mean,std = [],[]
  mean = [np.mean(data[:,:,:,i].reshape(-1)) for i in range(3)]
  std = [np.std(data[:,:,:,i].reshape(-1)) for i in range(3)]
  return np.array(mean),np.array(std)

train_x = np.load(os.path.join(path, 'trainX.npy'))
val_x = np.load(os.path.join(path, 'valX.npy'))
val_y = np.load(os.path.join(path, 'valY.npy'))
mean,std = cal_mean_std(np.concatenate([train_x, val_x]))
```

```python
'''
Dataset
'''
class AEDataset(data.Dataset):
  def __init__(self, path, name, with_labels = False, with_aug = True):
    self.path = path
    self.with_labels = with_labels
    self.with_aug = with_aug
    # 读取数据，初始化transformer
    self.init_data(name)
  
  def __getitem__(self,i):
    x = self.tsf(self.data[i])
    if self.with_labels:
      y = self.labels[i]
      return x # 这个任务比较特殊，所有阶段都不返回y
    return x

  def __len__(self):
    return len(self.data)

  def init_data(self, name):
    self.data = np.load(os.path.join(path, f'{name}X.npy'))
    if self.with_labels:
      self.labels = np.load(os.path.join(path, f'{name}Y.npy'))
    if self.with_aug:
      self.tsf = tfms.Compose([
          tfms.ToPILImage(),
          tfms.ColorJitter(brightness=0.15,contrast=0.15,saturation=0.15),
          # tfms.RandomHorizontalFlip(),
          tfms.RandomRotation(15),
          tfms.ToTensor(),
          tfms.Normalize(mean, std)
      ])
    else:
      self.tsf = tfms.Compose([
        tfms.ToPILImage(),
        tfms.ToTensor(),
        tfms.Normalize(mean, std)
      ])
```

```python
def get_dataloader(dataset, mode = 'train'):
  shuffle = True if mode == 'train' else False
  dataloader = data.DataLoader(dataset, batch_size = 64, shuffle = shuffle)
  return dataloader
```

```python
train_set = AEDataset(path, 'train', with_labels = False,with_aug = True)
val_set = AEDataset(path, 'val', with_labels = True, with_aug = False)
test_set = AEDataset(path, 'train', with_labels = False, with_aug = False)
```

## Model

Auto-Encoder模型，Encoder和Decoder相反，主要注意CNN的反向操作即可。

**`nn.ConvTranspose2d`**,这个是Pytorch的api，实际上不是真正的反卷积，但实现上可以使用。它的实现原理是先把input放大成新的特征图，再在大图上进行卷积操作。可以理解为做了Unpooling，再进行卷积。

实际使用的时候，只需要使用api提供的计算公式，计算出还原大小的kernel size即可。
> Hout =(Hin−1)×stride−2×padding+  kernel_size−1

```python
class AE(nn.Module):
    def __init__(self):
        super(AE, self).__init__()
        
        self.encoder = nn.Sequential(         # input 3 32 32
            nn.Conv2d(3, 64, 3, stride=1, padding=1),  # 64, 32, 32
            nn.ReLU(True),
            nn.MaxPool2d(2),                # 64, 16, 16
            nn.Conv2d(64, 128, 3, stride=1, padding=1), # 128, 16, 16
            nn.ReLU(True), 
            nn.MaxPool2d(2),                # 128, 8, 8
            nn.Conv2d(128, 256, 3, stride=1, padding=1),# 256, 8, 8
            nn.ReLU(True),
            nn.MaxPool2d(2)                 # 256, 4, 4
        )
 
        self.decoder = nn.Sequential(
            # out = (in - 1) * stride - 2 * padding + kernel_size
            nn.ConvTranspose2d(256, 128, 5, stride=1), # 128 8 8
            nn.ReLU(True),
            nn.ConvTranspose2d(128, 64, 9, stride=1), # 64 16 16 
            nn.ReLU(True),
            nn.ConvTranspose2d(64, 3, 17, stride=1),  # 3 32 32
            nn.Tanh() # 使用tanh保留完整值域
        )

    def forward(self, input):
        vec = self.encoder(input)
        x  = self.decoder(vec)
        return vec.view(len(input),-1), x
```

```python
model = AE()
print(summary(model,torch.zeros(1, 3, 32, 32), show_hierarchical=True))
```

## Train


```python
def run_epoch(model, dataloader, loss_func, optimizer, mode = 'train'):
  st = time.time()
  total_loss = 0
  vector = []
  x_rec = []
  if mode == 'train':
    for x_batch in dataloader:
      optimizer.zero_grad()
      x_batch = x_batch.to(device)
      vec,x = model(x_batch)
      loss = loss_func(x_batch, x)
      loss.backward()
      optimizer.step()
      total_loss += loss.item() * len(x_batch)
      vector.append(vec)
      x_rec.append(x)
  elif mode == 'eval':
    with torch.no_grad():
      for x_batch in dataloader:
        x_batch = x_batch.to(device)
        vec,x = model(x_batch)
        loss = loss_func(x_batch, x)
        total_loss += loss.item() * len(x_batch)
        vector.append(vec)
        x_rec.append(x)
  et = time.time()
  return torch.cat(vector,dim = 0), torch.cat(x_rec, dim = 0), total_loss/len(dataloader),et-st 
```

## Cluster

聚类，首先使用PCA降维，这是由于t-SNE计算开销太大。

```python
def cluster(vec, show_plot = False):
  # 选择最佳的维度
  pca = PCA()
  pca.fit(vec)
  cumsum = np.cumsum(pca.explained_variance_ratio_)
  # d = np.argmax(cumsum>0.99)+1
  d = 200
  if show_plot:
    print(d)
    plt.plot(cumsum)
    plt.show()
  tfs = KernelPCA(n_components=d, kernel='rbf')
  vec_1 = tfs.fit_transform(vec)
  # print('Fisrt Reduction Shape:{}'.format(vec_1.shape))
  vec_2 = TSNE(n_components=2, random_state=0).fit_transform(vec_1)
  # print("Second Reduction Shape:{}".format(vec_2.shape))
  labels = MiniBatchKMeans(2, random_state=0).fit(vec_2).labels_
  return vec_2, labels

def cal_acc(pred_labels, labels):
  return np.sum(pred_labels == labels)/len(labels)

def plot_scatter(feat, label, savefig=None):
    """ Plot Scatter Image.
    Args:
      feat: the (x, y) coordinate of clustering result, shape: (9000, 2)
      label: ground truth label of image (0/1), shape: (9000,)
    Returns:
      None
    """
    fig = plt.figure()
    X = feat[:, 0]
    Y = feat[:, 1]
    plt.scatter(X, Y, c = label)
    plt.legend(loc='best')
    if savefig is not None:
        plt.savefig(savefig)
    
    return fig

def predict(model, val_loader, show_plot = False):
  model.eval()
  latents,_,loss,_ = run_epoch(model, val_loader, loss_func, optimizer, mode= 'eval')
  vec = latents.detach().cpu().numpy()
  x_rd,labels = cluster(vec, show_plot=show_plot)
  return x_rd, labels, loss
```

```python
def plot_rec(model):
  # 畫出原圖
  plt.figure(figsize=(10,4))
  indexes = [1,2,3,6,7,9]
  imgs = test_set.data[indexes,]
  for i, img in enumerate(imgs):
      plt.subplot(2, 6, i+1, xticks=[], yticks=[])
      plt.imshow(img)
  # 运行model
  input_ = [test_set[i].unsqueeze(0) for i in indexes]
  input_ = torch.cat(input_,dim = 0).to(device)
  model.eval()
  with torch.no_grad():
    vec, x_rec = model(input_)
  # reconstruction
  x_rec = x_rec.permute(0,2,3,1).cpu().numpy()
  x_rec = (x_rec * std.reshape((1,1,1,3))+mean.reshape((1,1,1,3)))
  for i, img in enumerate(x_rec):
      plt.subplot(2, 6, 6+i+1, xticks=[], yticks=[])
      plt.imshow(img)
  plt.tight_layout()
```

### 512 dim latents

```python
class AE(nn.Module):
    def __init__(self):
        super(AE, self).__init__()
        
        self.encoder = nn.Sequential(         # input 3 32 32
            nn.Conv2d(3, 64, 3, stride=1, padding=1),  # 64, 32, 32
            nn.ReLU(True),
            nn.MaxPool2d(2),                # 64, 16, 16
            nn.Conv2d(64, 128, 3, stride=1, padding=1), # 128, 16, 16
            nn.ReLU(True), 
            nn.MaxPool2d(2),                # 128, 8, 8
            nn.Conv2d(128, 256, 3, stride=1, padding=1),# 256, 8, 8
            nn.ReLU(True),
            nn.MaxPool2d(2),                 # 256, 4, 4
            nn.Conv2d(256, 512, 3, stride=1, padding=1), # 512，4，4
            nn.ReLU(True),
            nn.MaxPool2d(2),                 # 512, 2, 2 
            nn.Conv2d(512, 512, 3, stride=1, padding=1), # 512, 2, 2 
            nn.ReLU(True),
            nn.MaxPool2d(2),                 # 512, 1, 1 
        )
 
        self.decoder = nn.Sequential(
            # out = (in - 1) * stride - 2 * padding + kernel_size
            nn.ConvTranspose2d(512, 512, 2, stride=1), # 256 2 2
            nn.ReLU(True),
            nn.ConvTranspose2d(512, 256, 3, stride=1), # 256 4 4
            nn.ReLU(True),
            nn.ConvTranspose2d(256, 256, 5, stride=1), # 256 8 8
            nn.ReLU(True),                  
            nn.ConvTranspose2d(256, 128, 1, stride=1), # 128 8 8
            nn.ReLU(True),
            nn.ConvTranspose2d(128, 64, 9, stride=1), # 64 16 16 
            nn.ReLU(True),
            nn.ConvTranspose2d(64, 3, 17, stride=1),  # 3 32 32
            nn.Tanh() # 使用tanh保留完整值域
        )

    def forward(self, input):
        vec = self.encoder(input)
        x  = self.decoder(vec)
        return vec.view(len(input),-1), x
```

```python
class AE(nn.Module):
    def __init__(self):
        super(AE, self).__init__()
        
        self.encoder = nn.Sequential(         # input 3 32 32
            nn.Conv2d(3, 64, 3, stride=1, padding=1),  # 64, 32, 32
            nn.ReLU(True),
            nn.MaxPool2d(2),                # 64, 16, 16
            nn.Conv2d(64, 128, 3, stride=1, padding=1), # 128, 16, 16
            nn.ReLU(True), 
            nn.MaxPool2d(2),                # 128, 8, 8
            nn.Conv2d(128, 256, 3, stride=1, padding=1),# 256, 8, 8
            nn.ReLU(True),
            nn.MaxPool2d(2),                 # 256, 4, 4
            nn.Conv2d(256, 256, 3, stride=1, padding=1),
            nn.ReLU(True),
            nn.MaxPool2d(2),                 # 256, 2, 2 
            nn.Conv2d(256, 256, 3, stride=1, padding=1),
            nn.ReLU(True),
            nn.MaxPool2d(2),                 # 256, 1, 1  
        )
 
        self.decoder = nn.Sequential(
            # out = (in - 1) * stride - 2 * padding + kernel_size
            nn.ConvTranspose2d(256, 256, 2, stride=1), # 256 2 2
            nn.ReLU(True),
            nn.ConvTranspose2d(256, 256, 3, stride=1), # 256 4 4
            nn.ReLU(True),
            nn.ConvTranspose2d(256, 256, 5, stride=1), # 256 8 8
            nn.ReLU(True),                  
            nn.ConvTranspose2d(256, 128, 1, stride=1), # 128 8 8
            nn.ReLU(True),
            nn.ConvTranspose2d(128, 64, 9, stride=1), # 64 16 16 
            nn.ReLU(True),
            nn.ConvTranspose2d(64, 3, 17, stride=1),  # 3 32 32
            nn.Tanh() # 使用tanh保留完整值域
        )
        self.fc = nn.Sequential(
            nn.Linear(256,2),
        )
    def forward(self, input):
        vec = self.encoder(input)
        x  = self.decoder(vec)
        vec = vec.view(len(input),-1)
        return self.fc(vec), x
```

```python
class AE(nn.Module):
    def __init__(self):
        super(AE, self).__init__()
        
        self.encoder = nn.Sequential(         # input 3 32 32
            nn.Conv2d(3, 64, 3, stride=1, padding=1),  # 64, 32, 32
            nn.ReLU(True),
            nn.MaxPool2d(2),                # 64, 16, 16
            nn.Conv2d(64, 128, 3, stride=1, padding=1), # 128, 16, 16
            nn.ReLU(True), 
            nn.MaxPool2d(2),                # 128, 8, 8
            nn.Conv2d(128, 256, 3, stride=1, padding=1),# 256, 8, 8
            nn.ReLU(True),
            nn.MaxPool2d(2),                 # 256, 4, 4
  
        )
 
        self.decoder = nn.Sequential(
            # out = (in - 1) * stride - 2 * padding + kernel_size
            nn.ConvTranspose2d(256, 256, 5, stride=1), # 256 8 8
            nn.ReLU(True),                  
            nn.ConvTranspose2d(256, 128, 1, stride=1), # 128 8 8
            nn.ReLU(True),
            nn.ConvTranspose2d(128, 64, 9, stride=1), # 64 16 16 
            nn.ReLU(True),
            nn.ConvTranspose2d(64, 3, 17, stride=1),  # 3 32 32
            nn.Tanh() # 使用tanh保留完整值域
        )
        self.fc = nn.Sequential(
            nn.Linear(4096,256),
            nn.ReLU(),
        )
        self.fc_rev = nn.Sequential(
            nn.Linear(256, 4096),
            nn.ReLU()
        )
    def forward(self, input):
        # encoder
        vec = self.encoder(input)
        shape = vec.size()
        vec = self.fc(vec.view(vec.size(0),-1))
        vec_rev = self.fc_rev(vec)
        # decoder
        x  = self.decoder(vec_rev.view(shape))

        return vec, x
```

```python
class AE(nn.Module):
    def __init__(self):
        super(AE, self).__init__()
        
        self.encoder = nn.Sequential(         # input 3 32 32
            nn.Conv2d(3, 64, 3, stride=1, padding=1),  # 64, 32, 32
            nn.BatchNorm2d(64),
            nn.ReLU(True),
            nn.MaxPool2d(2),                # 64, 16, 16
            nn.Conv2d(64, 128, 3, stride=1, padding=1), # 128, 16, 16
            nn.BatchNorm2d(128),
            nn.ReLU(True), 
            nn.MaxPool2d(2),                # 128, 8, 8
            nn.Conv2d(128, 256, 3, stride=1, padding=1),# 256, 8, 8
            nn.BatchNorm2d(256),
            nn.ReLU(True),
            nn.MaxPool2d(2)                 # 256, 4, 4
        )
 
        self.decoder = nn.Sequential(
            # out = (in - 1) * stride - 2 * padding + kernel_size
            nn.ConvTranspose2d(256, 128, 5, stride=1), # 128 8 8
            nn.BatchNorm2d(128),
            nn.ReLU(True),
            nn.ConvTranspose2d(128, 64, 9, stride=1), # 64 16 16 
            nn.BatchNorm2d(64),
            nn.ReLU(True),
            nn.ConvTranspose2d(64, 3, 17, stride=1),  # 3 32 32
            nn.BatchNorm2d(3),
            nn.Tanh() # 使用tanh保留完整值域
        )

    def forward(self, input):
        vec = self.encoder(input)
        x  = self.decoder(vec)
        return vec.view(len(input),-1), x
```
